```python
import pandas as pd
```

```python
nips = pd.read_csv('NIPS_1987-2015.csv')
nips = nips.loc[:,'1987_1':]
nips
```

```python
census = pd.read_csv('USCensus1990-2000rows.data.txt')
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
X = linkage(nips, 'ward')
```

```python
Y = linkage(census, 'ward')
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(X)
plt.title('NIPS_1987-2015')
plt.show()
```

```python
plt.figure()
dendrogram(Y)
plt.title('Census Data')
plt.show()
```
